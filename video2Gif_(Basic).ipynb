**VIDEO TO GIF**

Run the next cell, click **Upload** to upload a video, adjust settings, and click **Convert to GIF** to generate and download a gif.

```python
# @title
from google.colab import files
from moviepy.editor import VideoFileClip
import ipywidgets as widgets
from IPython.display import display, HTML
import os

def video_to_gif(input_video_path, output_gif_path, fps=10, scale=0.5):
    clip = VideoFileClip(input_video_path)
    clip = clip.resize(scale)
    clip.write_gif(output_gif_path, fps=fps)

def convert_and_download(video_file, fps, scale):
    output_gif_path = "converted_video.gif"
    video_to_gif(video_file, output_gif_path, fps=fps, scale=scale)
    files.download(output_gif_path)

upload_button = widgets.FileUpload(accept=".mp4,.mov,.avi", multiple=False)
fps_slider = widgets.IntSlider(value=10, min=1, max=30, step=1, description="FPS:")
scale_slider = widgets.FloatSlider(value=0.5, min=0.1, max=1.0, step=0.1, description="Scale:")
convert_button = widgets.Button(description="Convert to GIF")
output = widgets.Output()

def on_convert_button_click(change):
    with output:
        output.clear_output()

        if not upload_button.value:
            print("Please upload a video file first.")
            return

        uploaded_file_info = next(iter(upload_button.value.values()))
        video_file_name = uploaded_file_info.get('name', 'uploaded_video.mp4')
        with open(video_file_name, "wb") as f:
            f.write(uploaded_file_info.get('content', b""))

        try:
            convert_and_download(video_file_name, fps=fps_slider.value, scale=scale_slider.value)
        finally:
            if os.path.exists(video_file_name):
                os.remove(video_file_name)

convert_button.on_click(on_convert_button_click)

widgets_box = widgets.VBox([upload_button, fps_slider, scale_slider, convert_button, output])
display(widgets_box)
```
